```python
import matplotlib.pyplot as plt

# Assuming 'loss_table' and 'loss_table2' are two different DataFrames for your models.

fig, ax = plt.subplots(1, 2, figsize=(12, 6), sharey=True)  # 1 row, 2 columns of subplots

# Plot for the first model with a learning rate schedule
loss_table.plot(ax=ax[0], x='Step', 
                y=['Training Loss', 'Validation Loss', 'F1'],
                style={'Training Loss': '-', 'Validation Loss': '-', 'F1': '--'},
                title='Model with Learning Rate Schedule')

ax[0].set_xlabel('Step')
ax[0].set_ylabel('Value')
ax[0].axvline(x=500, color='black', linestyle='--', lw=2, label='Early Stopping Checkpoint')
ax[0].legend()
ax[0].legend().remove()

# Plot for the second model without a learning rate schedule
loss_table2.plot(ax=ax[1], x='Step', 
                 y=['Training Loss', 'Validation Loss', 'F1'],
                 style={'Training Loss': '-', 'Validation Loss': '-', 'F1': '--'},
                 title='Model without Learning Rate Schedule')

ax[1].set_xlabel('Step')
ax[1].axvline(x=700, color='black', linestyle='--', lw=2, label='Early Stopping Checkpoint')
ax[1].legend()
# Adjust the layout
plt.tight_layout()

# Since we want only one legend, we can remove the legend from the second subplot
ax[1].legend().remove()

# Place the legend on the right side of the last plot
# Use fig.legend() to create a shared legend at the bottom
handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=1)

# Adjust the subplot parameters to give some space for the legend
# plt.subplots_adjust(bottom=0.2)
plt.subplots_adjust(bottom=0.05)
# Save the figure
plt.savefig('combined_loss_plot.png', bbox_inches='tight')

# Show the plot
plt.show()
```
